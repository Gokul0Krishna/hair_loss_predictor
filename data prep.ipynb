```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('personality_dataset.csv')
```

```python
data.describe(include="all")
```

```python
null=data.isnull().any(axis=1).sum()
filled=(data.shape[0]-1)-null
plt.bar(['total','filled','null'], [data.shape[0]-1,filled,null],color=['b','c','r'])
plt.title('no. of null values comparition')
plt.xlabel('Fruits')
plt.show()
```

```python
x=data.isnull().sum()
```

```python
title=[]
da=[]
for  i in data:
    title.append(i.replace("_",' '))
for i in x:
    da.append(int(i))
print(title,da)
plt.barh(title,da)
plt.title('no. of null values per each column')
plt.xlabel('attributes')
plt.show()
```
